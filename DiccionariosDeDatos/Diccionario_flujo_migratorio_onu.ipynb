# Diccionario y Validación de Dataset migration_onu

Stock de migrantes internacionales 2020

Las estimaciones del número (o "stock") de migrantes internacionales desglosadas por edad, sexo y país o zona de origen se basan en estadísticas nacionales, en la mayoría de los casos obtenidas a partir de censos de población. Además, los registros de población y las encuestas representativas a nivel nacional proporcionan información sobre el número y la composición de los migrantes internacionales.

El conjunto de datos presenta estimaciones de migrantes internacionales desagregados por edad, sexo y origen. Se presentan estimaciones para 1990, 1995, 2000, 2005, 2010, 2015 y 2020 y están disponibles para 232 países y zonas del mundo. Las estimaciones se basan en estadísticas oficiales sobre los nacidos en el extranjero o la población extranjera.

Data extraida de: https://www.un.org/development/desa/pd/content/international-migrant-stock

Columnas encontradas inicialmente:

 - **Index:** Índice

 - **'Region, development group, country or area of destination':** "Región, grupo de desarrollo, país o zona de destino".

 - **Notes of destination:** Notas de destino.

 - **Location code of destination:** Código de localización del destino.

 - **Type of data of destination:** Tipo de datos de destino.

 - **Region.1**

 - **Development group.1:** grupo de desarrollo.1

 - **Country or area of origin:** país o zona de origen.

 - **Location code of origin:** Código de lugar de origen.
 
 - **Columnas númericas :** 1990	1995	2000	2005	2010	2015	2020 (se tomo en cuenta ambos sexos)
 - **Columnas númericas :** 1990.1	1995.1	2000.1	2005.1	2010.1	2015.1	2020.1 (se tomo en cuenta sólo hombres)
 - **Columnas númericas :** 1990.2	1995.2	2000.2 (se tomo en cuenta sólo mujeres)

### **Procesamiento de la data**

1. Debido a la estructura de nombres de los datos originales, los nombres de las columnas no coinciden con los valores. Por esta razón, les cambiamos el nombre para solucionar el problema:

    **onu.columns** = ['Index', 'Destination', 'Destination_Notes', 'Destination_Code', 'Destination_Type_of_Data', 'Origin', 'Origin_Code', '1990', '1995', '2000', '2005', '2010', '2015', '2020', '1990.1', '1995.1', '2000.1', '2005.1', '2010.1', '2015.1', '2020.1', '1990.2', '1995.2', '2000.2', '2005.2', '2010.2', '2015.2', '2020.2']
    
2. Eliminamos la columna "Índice" porque es innecesaria y repetitiva.
3. No necesitamos las columnas "Notas_de_destino" y "Tipo_de_datos_de_destino", por lo que también se eliminan.
4. Eliminamos los espacios en blanco de las columnas de años.
5. Los valores no existentes se representan con '..'. Tuvimos que cambiar eso.
6. Sustituimos los dos puntos por 0.
7. Aún queda algo de normalización por hacer. Algunas columnas tienen espacios en blanco y caracteres especiales.
8. Resolvemos el problema sustituyéndolos por nada, utilizando expresiones regulares.
9. Sabemos que los valores nulos están en dos columnas distintas. Vamos a buscarlos, podemos ver que la columna "Destino" tiene valores nulos para China y las regiones menos desarrolladas. El código de destino de China es 156, así que sustituimos los valores nulos por el código correcto.
10. Y sustituimos el código por el nombre del país en la columna "Origen".
11. Hicimos lo mismo con las regiones menos desarrolladas.
12. Hasta aquí sustituimos con éxito todos los valores nulos por los correctos.
13. La columna "Origin_Code" tiene valores correspondientes a la columna "Origin".
14. Creamos un diccionario con los valores de la columna 'Destino' como claves, y los valores de la columna 'Valores_Destino' como valores.
15. Ahora creamos una nueva columna llamada 'Origin_Code_New' y la rellenamos con los valores del diccionario, haciéndolos coincidir con los valores de la columna 'Origin'. De este modo, cada nombre de país y región coincidirá con su código correspondiente.
16. Sustituimos los valores de la columna "antiguo" "Origin_Code" por los valores de "Origin_Code_New". Y, por último, eliminamos "Origin_Code_New".
17. Buscamos duplicados y los eliminamos.
18. También eliminamos las filas que contienen los datos totales de migración, ya que afectarán a las estadísticas, y las filas que contienen valores que no podemos relacionar con un país, continente o región concretos.
19. Dividimos el conjunto de datos por pais, por region economica de origen, region economica de destino, region geografica de origen, region geografica de destino, continente de origen y continente de destino
20. Y los guardamos en formato .csv, son 7 datasets.
