We have no need for columns 'Destination_Notes' and 'Destination_Type_of_Data', therefore, they are also dropped

```python
onu.drop(['Destination_Notes', 'Destination_Type_of_Data'], axis=1, inplace=True)
```

```python
onu[['1990', '1995', '2000', '2005', '2010', '2015', '2020', '1990.1', '1995.1', '2000.1', '2005.1', '2010.1', '2015.1', '2020.1', '1990.2', '1995.2', '2000.2', '2005.2', '2010.2', '2015.2', '2020.2']].value_counts()
```

There's still some normalization to do. Some columns have white spaces and special characters.

```python
onu.Destination.value_counts()
```

We solve the problem by replacing them with nothing, using regular expressions

```python
onu = onu.replace('\W', '', regex=True)
```

We know the null values are in two distinct columns. Let's look for them

```python
onu.loc[onu['Destination_Code'].isna()]
```

As we can see, column 'Destination' has null values for China and Lessdevelopedregions, lets fix China first.

```python
onu.loc[onu['Destination'] == 'China']
```

We can see that the Destination code of China is 156, so we replace the null values with the correct code.

```python
onu.loc[onu['Destination'] == 'China'] = onu.loc[onu['Destination'] == 'China'].fillna(156)
```

Now we have to replace the code with the country's name in the column 'Origin'.

```python
onu.loc[onu['Destination'] == 'China']
```

```python
onu.Origin = onu.Origin.replace(156, 'China', regex=True)
```

```python
onu.loc[onu['Destination'] == 'China']
```

Now lets do the same for Lessdevelopedregions

```python
onu.loc[onu['Destination'] == 'Lessdevelopedregions']
```

```python
onu.loc[onu['Destination'] == 'Lessdevelopedregions'] = onu.loc[onu['Destination'] == 'Lessdevelopedregions'].fillna(902)
```

```python
onu.loc[onu['Destination'] == 'Lessdevelopedregions']
```

```python
onu.Origin = onu.Origin.replace(902, 'Lessdevelopedregions', regex=True)
```

```python
onu.loc[onu['Destination'] == 'Lessdevelopedregions']
```

We've successfuly replaced all null values with the correct ones

```python
onu.isna().sum()
```

Column 'Origin_Code' has values corresponding to column 'Origin'

```python
onu['Origin_Code'].unique()
```

We create a dictionary with the values of the column 'Destination' as keys, and the values of column 'Destination_Values as values.

```python
my_dict = pd.Series(onu.Destination_Code.values,index=onu.Destination).to_dict()
my_dict
```

Now we create a new column called 'Origin_Code_New' and fill it with the values of the dictionary, matching them to the values of column 'Origin'. This will match each country and region name to it's corresponding code.

```python
onu['Origin_Code_New'] = onu.Origin.replace(my_dict)
onu
```

We replace the values of the 'old' 'Origin_Code' column to the values of 'Origin_Code_New'

```python
onu.Origin_Code = onu.Origin_Code_New
```

And finally, we drop 'Origin_Code_New'

```python
onu.drop(['Origin_Code_New'], axis=1, inplace=True)
```
