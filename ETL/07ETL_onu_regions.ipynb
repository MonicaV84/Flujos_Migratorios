```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
regions_by_destination = pd.read_csv('..\\DataSets\\ONU\\regions_destination.csv')
regions_by_destination.head()
regions_by_destination.drop('Unnamed: 0', axis=1, inplace=True)
regions_by_destination.head()
```

```python
regions_by_origin = pd.read_csv('..\\DataSets\\ONU\\regions_origin.csv')
regions_by_origin.head()
regions_by_origin.drop('Unnamed: 0', axis=1, inplace=True)
regions_by_origin.head()
```

```python
destinations = regions_by_destination.loc[(regions_by_destination['Destination'] == 'EuropeandNorthernAmerica') | (regions_by_destination['Destination'] == 'LatinAmericaandtheCaribbean') | (regions_by_destination['Destination'] == 'EasternandSouthEasternAsia') | (regions_by_destination['Destination'] == 'SubSaharanAfrica')]
destinations
```

```python
plt.figure(figsize=(14, 8))
plt.ticklabel_format(style='plain', axis='y',useOffset=False)
sns.barplot(
    y="1990", 
    x="Destination", 
    data=destinations, 
    estimator=sum, 
    errorbar=None, 
    color='#69b3a2'
    )
plt.show()
```
