```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
destination_by_income = pd.read_csv('..\\DataSets\\ONU\\regions_by_income_destination.csv')
destination_by_income.head()
```

```python
len(destination_by_income['Destination'].unique())
```

```python
origin_by_income = pd.read_csv('..\\DataSets\\ONU\\regions_by_income_origin.csv')
origin_by_income.head()
```

```python
destination_by_income = destination_by_income.drop(destination_by_income[(destination_by_income['1990'].str.len() > 10) | (destination_by_income['1990'] == 'MacaoSAR')].index)
```

```python
destination_by_income['1990']= pd.to_numeric(destination_by_income['1990'])
```

```python
destination_1990 = destination_by_income.groupby(['Destination'])['1990'].sum()
destination_1990
```

```python
sns.set(font_scale=1.4)
destination_1990.plot(figsize=(22, 8), linewidth=2.5, color='green')
plt.xlabel("Destination")
plt.ylabel("Number of migrants in the year 1990")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Migrants per economical region of destination 1990", y=1.02, fontsize=22)
plt.show()
```
