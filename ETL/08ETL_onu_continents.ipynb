```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
continents_destination = pd.read_csv('..\\DataSets\\ONU\\continents_destination.csv')
continents_destination.head()
```

```python
continents_destination.info()
```

```python
destination_1990 = continents_destination.groupby(['Destination'])['1990'].sum()
destination_1990
```

```python
sns.set(font_scale=1.4)
destination_1990.plot(figsize=(22, 8), linewidth=2.5, color='green')
plt.xlabel("Destination")
plt.ylabel("Number of migrants in the year 1990")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Migrants per continent of destination 1990", y=1.02, fontsize=22)
plt.show()
```
